The Law of Large Numbers (LLN) states that as the number of observations or trials increases, the average of the observed values will converge to the expected value of the underlying distribution. This means that as the sample size increases, the sample mean will approach the population mean.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the sample size (number of dice rolls)
n = 1000

# Simulate rolling a fair six-sided die n times and calculate the mean
dice_rolls = np.random.randint(low=1, high=7, size=n)
sample_mean = np.mean(dice_rolls)
print(sample_mean)

# Repeat the above process for multiple sample sizes
sample_sizes = [10, 50, 100, 500, 1000, 5000, 10000]
means = []
for size in sample_sizes:
    rolls = np.random.randint(low=1, high=7, size=size)
    mean = np.mean(rolls)
    means.append(mean)

# Plot the sample means against sample sizes
plt.plot(sample_sizes, means)
plt.axhline(y=3.5, color='r', linestyle='--')
plt.xlabel('Sample Size')
plt.ylabel('Sample Mean')
plt.title('Law of Large Numbers')
plt.show()
```
